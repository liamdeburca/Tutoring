```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 1000, endpoint=True)

def x(t):
    return np.cos(t)

def y(t):
    return np.sin(t)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,4))
ax.set_title('Sinus og Cosinus')

ax.plot(t, x(t), label='x(t) = cos(t)')
ax.plot(t, y(t), label='y(t) = sin(t)')
ax.plot([t[0], t[-1]], [0, 0], c='k')

ax.plot([np.pi/2, np.pi/2], [-1.1, 1.1], c='k', linestyle='dotted')
ax.plot([np.pi, np.pi], [-1.1, 1.1], c='k', linestyle='dotted')
ax.plot([3*np.pi/2, 3*np.pi/2], [-1.1, 1.1], c='k', linestyle='dotted')

ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['$t=0$', '$t=\\pi/2$', '$t=\\pi$', '$t=3\\pi/2$', '$t=2\\pi$'])
ax.set_yticks([-1, 0, 1], [-1, 0, 1])

ax.set_xlim(t[0], t[-1])
ax.set_ylim(-1.1, 1.1)

plt.legend()
plt.savefig('figures/sinus_cosinus.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.set_title('Enhedscirkel')

ax.plot(x(t), y(t))

ax.plot([-1.1, 1.1], [0, 0], c='k')
ax.plot([0, 0], [-1.1, 1.1], c='k')

for frac, frac_str in zip([0, 1/4, 1/2, 3/4, 1, 5/4, 3/2, 7/4], ['0', '1/4', '1/2', '3/4', '1', '5/4', '3/2', '7/4']):

    angle = frac * np.pi
    degree = np.rad2deg(angle)

    if frac == '0':
        label = f'$t=0=0\\degree$'
    else:
        label = f'$t={frac_str}\\pi={int(degree)}\\degree$'

    ax.scatter([x(angle)], [y(angle)])
    ax.annotate(label, (x(angle), y(angle)-0.05))

ax.set_xticks([-1, -3**0.5/2, -0.5**0.5, -0.5, 0, 0.5, 0.5**0.5, 3**0.5/2, 1], ['$-1$', '$-\\sqrt{3}/2$', '$-1/\\sqrt{2}$', '$-0.5$', '$0$', '$0.5$', '$1/\\sqrt{2}$', '$\\sqrt{3}/2$', '$1$'], rotation=-90)
ax.set_yticks([-1, -3**0.5/2, -0.5**0.5, -0.5, 0, 0.5, 0.5**0.5, 3**0.5/2, 1], ['$-1$', '$-\\sqrt{3}/2$', '$-1/\\sqrt{2}$', '$-0.5$', '$0$', '$0.5$', '$1/\\sqrt{2}$', '$\\sqrt{3}/2$', '$1$'])
ax.grid()

ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)

ax.set_xlabel('x=cos(t)')
ax.set_ylabel('y=sin(t)')

plt.savefig('figures/enhedscirkel.png', dpi=300, bbox_inches='tight')
plt.show()
```
