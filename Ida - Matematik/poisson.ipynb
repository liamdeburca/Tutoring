```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import binom, poisson
```

Poisson fordelingen er en udvikling af binomialfordelingen, hvor antallet af forsøg, $n$, stiger samtidigt med at sandsynligheden for succes, $p$, falder. Dvs. forventningsværdien er konstant, derfor:

$$\text{Forventningsværdi: }\lambda$$
$$\text{Sandsynlighed: }p=\frac{\lambda}{n}$$

```python
l = 5 # Forventningsværdi
ks = np.arange(0, 10.1, 1, dtype=int) # x-akse, dvs. forskellige antal succeser
ns = [10, 25, 100, 1000, 10_000] # Liste med antal forsøg

fig, axes = plt.subplots(len(ns), 1, figsize=(12, 4*len(ns)), dpi=300)
for n, ax in zip(ns, axes):
    prob = l / n # Sandsynlighed for succes

    b = binom(n, prob)
    p = poisson(l)

    ax.set_title(f"{n} forsøg -> p = {prob*100}%")
    ax.plot(ks, b.pmf(ks), '-o', color='k', label='binomial')
    ax.plot(ks, p.pmf(ks), '-o', color='r', label='poisson')
    ax.set_xlim(0, ks[-1])
    ax.set_ylim(0)
    ax.legend()

plt.show()
```
