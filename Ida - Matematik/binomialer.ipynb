```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
def plot_rngs(rng, m=None):
    xmin = rng.min()-0.5
    xmax = rng.max()+1.5
    n_bins = (xmax - xmin)
    bins = np.linspace(xmin, xmax, int(n_bins)+1, endpoint=True)

    fig, ax = plt.subplots(figsize=(12,4), dpi=300)

    ax.hist(rng, bins=bins, color='navy')

    if m is not None:
        ax.scatter([m], [0], c='r', s=100)

    ax.set_xlim(xmin, xmax); ax.set_ylim(0)
    ax.set_xlabel('r'); ax.set_ylabel('Frekvens')

    plt.tight_layout()
    plt.show()
```

```python
N = 10

n1, p1 = 8, 0.4    # 3.2
n2, p2 = 23, 0.75  # 17.25
n3, p3 = 100, 0.55 # 55

rng1 = binom.rvs(n1, p1, size=N)
rng2 = binom.rvs(n2, p2, size=N)
rng3 = binom.rvs(n3, p3, size=N)

plot_rngs(rng1, m=3.2)
plot_rngs(rng2, m=17.25)
plot_rngs(rng3, m=55)
```
